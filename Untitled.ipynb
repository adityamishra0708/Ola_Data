```python
df['datetime'] = pd.to_datetime(df['datetime'])
```

```python
df.dtypes
```

```python
total_rides = df['count'].sum()
total_casual = df['casual'].sum()
total_registered = df['registered'].sum()
labels = ['Casual', 'Registered']
sizes = [total_casual, total_registered]
plt.pie(sizes, labels=labels,autopct='%1.1f%%')
plt.title('User Type Breakdown')
plt.axis('equal')
plt.show()
```

```python
daily_rides = df.groupby(df['datetime'].dt.date)['count'].sum()
plt.figure(figsize=(12, 4))
daily_rides.plot()
plt.title('Total Daily Rides Over Time')
plt.xlabel('Date')
plt.ylabel('Ride Count')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['month'] = df['datetime'].dt.to_period('M')
monthly_rides = df.groupby('month')['count'].sum()
plt.figure(figsize=(10, 4))
monthly_rides.plot(kind='bar', color='skyblue')
plt.title('Total Monthly Rides')
plt.xlabel('Month')
plt.ylabel('Ride Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['hour'] = df['datetime'].dt.hour
hourly_avg = df.groupby('hour')['count'].mean()
plt.figure(figsize=(8, 4))
hourly_avg.plot(kind='bar', color='lightgreen')
plt.title('Average Hourly Ride Demand')
plt.xlabel('Hour of Day')
plt.ylabel('Average Ride Count')
plt.xticks(range(0, 24),rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
weather_map = {
    1: 'Clear',
    2: 'Mist/Cloudy',
    3: 'Light Snow/Rain',
    4: 'Heavy Snow/Rain'
}
df['weather_desc'] = df['weather'].map(weather_map)
```

```python
weather_avg = df.groupby('weather_desc')['count'].mean().sort_values()
plt.figure(figsize=(8, 4))
sns.barplot(x=weather_avg.index, y=weather_avg.values, palette='coolwarm')
plt.title('Average Ride Count by Weather')
plt.ylabel('Average Ride Count')
plt.xlabel('Weather Condition')
plt.xticks(rotation=15)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
correlation = df[['temp', 'humidity', 'windspeed', 'count']].corr()
print(correlation['count'].sort_values(ascending=False))
```

```python
hourly_avg = df.groupby('hour')['count'].mean()
plt.figure(figsize=(10, 4))
sns.barplot(x=hourly_avg.index, y=hourly_avg.values, color='skyblue')
plt.title('Average Ride Count by Hour of Day')
plt.xlabel('Hour (023)')
plt.ylabel('Average Ride Count')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
hourly_user_avg = df.groupby('hour')[['casual', 'registered']].mean().reset_index()
plt.figure(figsize=(10, 4))
sns.lineplot(x='hour', y='casual', data=hourly_user_avg, label='Casual', marker='o')
sns.lineplot(x='hour', y='registered', data=hourly_user_avg, label='Registered', marker='o')
plt.title('Hourly Ride Pattern: Casual vs Registered')
plt.xlabel('Hour of Day')
plt.ylabel('Average Ride Count')
plt.grid(True)
plt.xticks(range(0, 24))
plt.legend()
plt.tight_layout()
plt.show()
```
